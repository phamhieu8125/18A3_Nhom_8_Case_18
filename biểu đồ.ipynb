```python
import matplotlib.pyplot as plt
import pandas as pd

event = pd.read_csv(r'C:\Users\DELL\18A3_Nhom_8_Case_18\data_clean\event_attendance_clean.csv')
event['checkin_time'] = pd.to_datetime(event['checkin_time'], errors = 'coerce')
event_time = event.dropna(subset= ['checkin_time']).copy()
event_time['month'] = event_time['checkin_time'].dt.to_period('M').astype(str)
counts_per_month = event_time.groupby('month').size().sort_index()
x = counts_per_month.index
y = counts_per_month.values

plt.figure(figsize= (10, 4))
plt.plot(x, y, marker = 'o')
plt.title('Số lượng check_in theo tháng')
plt.xlabel('Tháng (YYYY-MM)')
plt.ylabel('Số lượt điểm danh')
plt.grid(True)
plt.xticks(rotation = 45)
plt.show
```
