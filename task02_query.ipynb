```python
import pandas as pd

student = pd.read_csv('data_clean/student_profile_clean.csv')
event   = pd.read_csv('data_clean/event_attendance_clean.csv')
club    = pd.read_csv('data_clean/club_registration_clean.csv')

```

```python
student_df = pd.DataFrame(student)
event_df = pd.DataFrame(event)
club_df = pd.DataFrame(club)
```

```python
print("Thống kê số lượng sinh viên theo class_cohort")
loc_theo_nhom_lop = (
    student_df
    .groupby("class_cohort")["student_id"]
    .nunique()
    .sort_index()
)
print(loc_theo_nhom_lop)
```

```python
print("Thống kê số lượng sinh viên đăng ký mỗi CLB (dựa trên club_registration)")
loc_theo_nhom = (
    club_df
    .groupby("club_id")["student_id"]
    .nunique()
    .sort_index()
)
print(loc_theo_nhom)
```

```python
print("Thống kê số lượng sự kiện theo CLB")
loc_theo_su_kien = (
    event_df
    .groupby("club_id")["event_id"]
    .nunique()

)
print(loc_theo_su_kien)
```

```python
print("Thống kê số lượt tham gia sự kiện theo attendance_status.")
loc_theo_su_kien = (
    event_df
    .groupby("attendance_status")["event_id"]
    .nunique()
    .sort_index()
)
print(loc_theo_su_kien)
```

```python
ban_ghi_khong_hop_le = event_df[
    (~event_df["student_id"].isin(student_df))|
    (~event_df["club_id"].isin(club_df))]
print("Các bản ghi không hợp lệ trong event_attendance:")
print(ban_ghi_khong_hop_le)
```
