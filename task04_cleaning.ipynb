```python
import pandas as pd #Đọc file dữ liệu
student = pd.read_csv('data_clean/student_profile_clean.csv')
event   = pd.read_csv('data_clean/event_attendance_clean.csv')
club    = pd.read_csv('data_clean/club_registration_clean.csv')

student_club = student.merge( #Gộp theo student id và phân tích số clb theo lớp khóa
    club_reg,
    on="student_id",
    how="left"
)

event_full = event.merge(
    student,
    on="student_id", #Merger theo student id
    how="left"
)

event_full = event_full.merge(
    club_reg,
    on=["student_id", "club_id"], #Merge theo student id với club id
    how="left"
)
sv_khong_tham_gia_club = event_full.loc[
    event_full["club_id"].isna() 
]

reg_count = club_reg.groupby("club_id").agg( #Số người tham gia clb
    nguoi_dang_ki=("student_id", "nunique")
)

att_count = event.groupby("club_id").agg(
    attended=("student_id", "nunique")
)


thong_ke_clb = reg_count.merge(
    att_count,
    on="club_id",
    how="left"
).fillna(0)

thong_ke_clb["attendance_rate"] = (
    thong_ke_clb["attended"] / thong_ke_clb["registered"]
)
clb_ty_le_tham_gia_thap = thong_ke_clb.loc[ #tỷ lệ tham gia clb thấp
    thong_ke_clb["attendance_rate"] < 0.3
]
```
