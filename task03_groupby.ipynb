```python
import pandas as pd

df_sinh_vien = pd.read_csv('data_clean/student_profile_clean.csv')
df_dang_ky_clb = pd.read_csv('data_clean/club_registration_clean.csv')
df_tham_gia_sk = pd.read_csv('data_clean/event_attendance_clean.csv')
```

```python
#Số lần 
merge_clb_sv = pd.merge(df_dang_ky_clb, df_sinh_vien, on = 'student_id', how = 'left')

tb_clb_theo_lop = merge_clb_sv.groupby('class_cohort')['join_count'].mean().reset_index()
tb_clb_theo_lop.columns = ['Lớp/Khóa', 'TB_Tham_Gia_CLB']

print("=== 1. Số lần tham gia CLB trung bình theo class_cohort ===")
print(tb_clb_theo_lop.sort_values(by = 'TB_Tham_Gia_CLB', ascending = False).head())
```

```python
so_sv_dang_ky = df_dang_ky_clb.groupby('club_id')['student_id'].count().reset_index()
so_sv_dang_ky.columns = ['Mã CLB', 'Số_SV_Đăng_Ký']

print("\n=== 2. Số sinh viên đăng ký theo club_id ===")
print(so_sv_dang_ky.sort_values(by = 'Số_SV_Đăng_Ký', ascending = False).head())
```

```python
df_co_mat = df_tham_gia_sk[df_tham_gia_sk['attendance_status'] == 'Present']

so_luot_co_mat_theo_clb = df_co_mat.groupby('club_id')['student_id'].count()

print("\n===  3. Số lượt tham gia sự kiện (attendance_status = 'Present') theo club_id ===")
print(so_luot_co_mat_theo_clb.sort_values(ascending = False))
```

```python
df_merge_sk_sv = pd.merge(df_tham_gia_sk, df_sinh_vien, on = 'student_id', how = 'left')

so_sk_cua_moi_sv = df_merge_sk_sv.groupby(['student_id', 'class_cohort']).size().reset_index(name = 'so_lan_tham_gia')
trung_binh_sk_theo_lop = so_sk_cua_moi_sv.groupby('class_cohort')['so_lan_tham_gia'].mean()

print("\n=== 4. Số lượt tham gia sự kiện trung bình theo class_cohort ===")
print(trung_binh_sk_theo_lop.sort_values(ascending = False))
```
