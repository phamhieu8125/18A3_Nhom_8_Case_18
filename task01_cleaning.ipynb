2. Làm sạch và chuẩn hóa từng file csv

2.1 Làm sạch và chuẩn hóa file student_profile.csv

```python
# a) student_id
df_student['student_id'] = df_student['student_id'].str.strip().str.upper()
```

Ý nghĩa: làm sạch (clean) dữ liệu cột student_id bằng cách:-Xóa khoảng trắng dư ở đầu và cuối chuỗi
                                                           -Chuyển toàn bộ ký tự sang chữ in hoa
Giải thích:
df_student['student_id']: Truy cập cột student_id trong DataFrame df_student,kiểu dl: pandas.Series,p.tử thường là chuỗi (string)
df_student['student_id'].str.strip().str.upper():
-df_student['student_id'].str: dùng hàm xử lý chuỗi toàn bộ Series(không có .str gây ra lỗi)
-.str.strip():xóa khoảng cách đầu cuối chuỗi
-.str.upper():chuyển các chữ cái thành chữ in hoa
--> cơ chế hoạt động các bước:
-Lấy cột student_id
-strip() : xóa khoảng trắng
-upper() : in hoa
-Gán lại vào cột student_id

```python
# b) full_name
df_student['full_name'] = (
    df_student['full_name']
    .str.strip()
    .str.split()
    .str.join(' ')
    .str.title()
)
```

Ý nghĩa: chuẩn hóa cột full_name (họ và tên sinh viên) bằng cách:-Xóa khoảng trắng dư ở đầu, cuối
                                                                 -Gộp nhiều khoảng trắng liên tiếp thành 1 khoảng trắng
                                                                 -Viết hoa chữ cái đầu mỗi từ (Title Case)
Giải thích: 
df_student['full_name']:Truy cập cột full_name trong DataFrame df_student,kdl: pandas.Series,chuỗi ký tự (string)
(df_student['full_name'].str.strip()
                           .str.replace(r'\s+', ' ', regex=True)
                           .str.title()):
-.str:String accessor của pandas,dùng các hàm xử lý chuỗi cho toàn bộ Series,không có .str gây ra lôi
-.str.strip(): xóa khoảng trắng đầu cuối
-.str.split(): Tách chuỗi thành danh sách các từ.Tách theo mọi loại khoảng trắng:space;tab;newline.Tự động bỏ khoảng trắng dư
-.str.join(' '): Ghép các từ vừa tách thành 1 chuỗi; dùng ' ' nghĩa là 1 dấu cách để nối
-.str.title(): viết hoa các chữ cái đầu mỗi từ các chữ cái còn lại thanh chữ viết thường

```python
# c) class_cohort
df_student['class_cohort'] =df_student['class_cohort'].str.strip().str.split().str.join('').str.upper()
```

Ý nghĩa: chuẩn hóa cột class_cohorts(tên lớp) bằng cách: -xóa khoảng trắng dư ở đầu, cuối
                                                         -viết in hoa các kí tự bao gồm chữ và số
Giải thích:
-df_student['class_cohort']:Truy cập cột class_cohort trong DataFrame df_student,kdl: pandas.Series,chuỗi ký tự (string)
-.str: không có sẽ bị lỗi vì các hàm như .strip();.upper() không phải là hàm của Series,mà là hàm của kiểu chuỗi (string)
       (để nói với pandas rằng:Hãy áp dụng hàm chuỗi này cho từng phần tử trong cột)
-.str.strip(): xóa khoản trắng đầu cuối
-.str.upper(): in hoa toàn bộ kí tự

2.2 Làm sạch và chuẩn hóa club_registration.csv

```python
# a) student_id
df_club['student_id'] = df_club['student_id'].str.strip().str.upper()
```

```python
# b) club_id
df_club['club_id'] = (
    df_club['club_id']
    .astype(str)
    .str.strip()
    .str.upper()
    .str.replace(r'CLB', '', regex=True)
    .str.replace(r'[-\s]+', '_', regex=True)
    .str.replace(r'[^A-Z0-9_]', '', regex=True)
    .str.strip('_')
    .radd('CLB_')
)
```

Ý nghĩa và giải thích: giống với ý 2.1c chỉ khác truy cập vào cột club_id trong DataFrame df_club

```python
# c) join_count
df_club['join_count'] = (
    pd.to_numeric(df_club['join_count'], errors='coerce')
    .fillna(0)
    .clip(lower=0)
)
```

Ý nghĩa: -luôn là số, không Nan, không số âm.
Ý tưởng: đổi hết sang số nếu có lỗi thì sẽ thành Nan -> Nan sẽ thay bằng 0 -> Số dương giữ nguyên, số âm thành 0
Giải thích:
-pd.to_numeric(df_club['join_count'], errors='coerce'): chuyển cột jojn_count trong df_club sang số, 
                                                        errors='coerce': thay vì báo lỗi sẽ báo Nan
-.fillna(0): chuyển Nan thành 0
-.clip(lower=0): dương giữ, âm chuyển thành 0

2.3 Làm sạch và chuẩn hóa event_attendance.csv

```python
# a) event_id, studen_id, club_id
df_event[['event_id', 'student_id', 'club_id']] = (
    df_event[['event_id', 'student_id', 'club_id']]
    .apply(lambda x: x.str.strip().str.upper())
)
```

Ý nghĩa: Chuẩn hóa 3 cột ID (event_id, student_id, club_id) bằng cách:-Xóa khoảng trắng thừa ở đầu và cuối
                                                                      -Chuyển ký tự thành chữ in hoa
Giải thích: 
-df_event[['event_id', 'student_id', 'club_id']]: lấy 3 cột event_id, student_id, club_id trong df_event để chuẩn hóa(kdl: DataFrame ko phải Series)
-.apply(...): Áp dụng một hàm cho từng cột (mặc định axis=0)
              Mỗi lần chạy:x = 1 cột (kiểu pandas.Series)
-lambda x: lamda = hàm ẩn danh
           lamda x = def clean_column(x):return x.str.strip().str.upper()
-x.str.strip().str.upper(): chuẩn hóa lại dữ liệu xóa bỏ khoảng trắng thừa đầu cuối và chuyển kí tự thành chữ in hoa
-df_event[['event_id', 'student_id', 'club_id']] = : sau khi chuẩn hóa và làm sạch thì ghi đè lên 

```python
# b) event_name
df_event['event_name'] = (
    df_event['event_name']
    .str.strip()
    .str.split()
    .str.join(' ')
    .str.title()
)
```

Ý nghĩa và giải thích: giống như câu 2.1b

```python
# c)
df_event['club_id'] = (
    df_event['club_id']
    .astype(str)
    .str.strip()
    .str.upper()
    .str.replace(r'CLB', '', regex=True)
    .str.replace(r'[-\s]+', '_', regex=True)
    .str.replace(r'[^A-Z0-9_]', '', regex=True)
    .str.strip('_')
    .radd('CLB_')
)

```

```python
# d) attendance_status
status_map = {
    'CO MAT': 'Present',
    'CÓ MẶT': 'Present',

    'VANG': 'Absent',
    'VẮNG': 'Absent',
    
    'MUON': 'Late',
    'MUỘN': 'Late',
}

df_event['attendance_status'] = (
    df_event['attendance_status']
    .astype(str)
    .str.strip()
    .str.upper()
    .str.split()
    .str.join(' ')
    .map(status_map)
)

```

Ý nghĩa: chuẩn hóa toàn bộ cột attendance_status về 1 dạng thống nhất bao gồm (Present, Absent, Late và Unknow)
df_event['attendance_status'] = (
    df_event['attendance_status']
    .astype(str) : ép toàn bộ giá trị trong cột thành chuỗi để không bị lỗi nếu dùng .str  
    .str.strip() : xóa khoảng trắng dư thừa đầu và cuối chuỗi
    .str.upper() : in hoa toàn bộ kí tự
    .str.split() : tách chuỗi thành danh sách các từ.Tách theo mọi loại khoảng trắng:space;tab;newline.Tự động bỏ khoảng trắng dư
    .str.join(' ') : ghép các phần tử vừa tách thành 1 chuỗi, lấy ' ' là 1 khoảng cách để nối
    .map(status_map) : lấy từng giá trị tra trong bảng ánh xạ status_map thay bằng giá trị chuẩn, không có dữ liệu thì sẽ thành Nan
    .fillna('Unknow'): Nan --> Unknow (nếu muốn có thể thay sang thành Absent)
)

```python
# d) checkin_time
df_event['checkin_time'] = (
    df_event['checkin_time']
    .astype(str)
    .str.strip()
    .str.replace(r'\s+', ' ', regex=True)
)

df_event['checkin_time'] = pd.to_datetime(
    df_event['checkin_time'],
    errors='coerce',
    infer_datetime_format=True
)

```
